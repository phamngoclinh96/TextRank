## Đồ thị biểu diễn liên kết các từ

```python
import networkx
%matplotlib inline
import matplotlib
```

```python
def build_graph(text):
    _g = get_graph(text)
    _graph = networkx.Graph()
    for edge in _g.edges():
        _graph.add_edge(*edge)
    return _graph
```

```python
graph = build_graph(text)
matplotlib.rcParams['figure.figsize'] = (20.0, 16.0)
networkx.draw_networkx(graph)
```
